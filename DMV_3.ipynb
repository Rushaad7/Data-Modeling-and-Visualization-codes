Check for inconsistent data, such as inconsistent formatting or spelling variations and standardize it.
Convert columns to the correct data types as needed.

```python
df['TotalCharges'].replace(" ",0,inplace=True)
```

```python
df['TotalCharges']=df['TotalCharges'].astype(float)  #converting total charges to float
```

```python
df.describe()
```

```python
objcols=[]
for column in df.select_dtypes(include='object').columns:
    objcols.append(column)
    print(column,df[column].unique(),len(df[column].unique()))
```

Here we can see that the following set of columns are categorical in nature with 2-4 unique categories in each column. These can be converted to numerical values using one-hot encoding or label encoding so that it can be used for customer churn prediction

As all columns have ordinal data we will perform label encoding of the data

```python
label_encoders={}
# label_encodes=LabelEncoder()
for col in objcols:
    label_encoders[col] = LabelEncoder()
    df[col] = label_encoders[col].fit_transform(df[col])

# label_encoders
```

```python
df.head()
```

Handling Outliers

```python
z_scores = zscore(df.select_dtypes(include=['float64', 'int64']))
```

```python
z_scores
```

```python
outliers_removed = df[(z_scores < 3).all(axis=1)]
```

```python
outliers_removed
```

Feature Engineering

```python
df['MonthlyToTotalRatio'] = df['MonthlyCharges'] / df['TotalCharges']
```
